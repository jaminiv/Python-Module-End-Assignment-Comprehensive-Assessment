```python
#Analysis Tasks:
#1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)

team=Employee["Team"].value_counts()
total_employees = len(Employee)
percentage_split = (team / total_employees) * 100
print(percentage_split)

plt.figure(figsize=(10, 6))
team.plot(kind='bar', color='indigo')
plt.xlabel('Team')
plt.ylabel('Percentage split')
plt.title('Percentage')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
#Insight: Here we find out there 30 teams and their percentiles are as given below:
```

```python
#2. Segregate employees based on their positions within the company. (2 marks)

pt=Employee.sort_values(by=["Position"], ascending=True)
print(pt)
position_counts = Employee['Position'].value_counts()
plt.pie(position_counts,labels=position_counts.index,autopct="%1.1f%%")
plt.title("Employee Distribution by Position")
plt.show()
#Insight: By segregating the employee on basis of 'Position', we come to a conclusion that there are more employees from the Position SG.
```

```python
#3. Identify the predominant age group among employees. (2 marks)

age=Employee['Age'].value_counts()
index_age = Employee['Age'].value_counts().idxmax() #index max
max_age = Employee['Age'].value_counts().max() #max number of times

print("\nThe predominant age group amoung employees is",index_age,".Total number of people with the same age group is ",max_age)
plt.bar(age.index, age.values,color="maroon")
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.title('Employee Distribution by Age Group')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
#Insight: Most of the employees hired, are hired at the age of 24 years.
```

```python
#4. Discover which team and position have the highest salary expenditure. (2 marks)

saltab = pd.DataFrame(Employee)
grouped_data = saltab.groupby(['Team','Position'])
total_salary_by_group = grouped_data['Salary'].sum()
highest_expenditure_group = total_salary_by_group.idxmax()
highest_spending_team = highest_expenditure_group[0]
highest_spending_position = highest_expenditure_group[1]
print(f"Team with Highest Salary Expenditure: {highest_spending_team}")
print(f"Position with Highest Salary Expenditure: {highest_spending_position}")
total_salary_by_group.plot(kind='line',color='maroon')
plt.xlabel('Team,Position')
plt.ylabel('Total Salary')
plt.title('Salary Expenditure')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#Insight: In this we can conclude that Los Angles Lakers team is getting higher paid than other teams and the position getting higher paid is SF. 
```
